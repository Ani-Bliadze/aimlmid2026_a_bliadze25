```python
import math
import matplotlib.pyplot as plt
import numpy as np


x = [-8.52, -5.1, -2.7, -0.4, 1.8, 4.2, 7.04, 9.1]
y = [-6.76, -3.5, -1.4, 0.7, 2.6, 4.8, 6.3, 8.1]

def mean(arr):
    return sum(arr) / len(arr)

def covariance(arr1, arr2, mean_x, mean_y):
    return sum((arr1[i] - mean_x) * (arr2[i] - mean_y) for i in range(len(arr1)))

def sigma(arr, m):
    return math.sqrt(sum((arr[i] - m) ** 2 for i in range(len(arr))))

def calculate_pearson(x, y):
    mx, my = mean(x), mean(y)
    num = covariance(x, y, mx, my)
    den = sigma(x, mx) * sigma(y, my)
    return num / den

r_value = calculate_pearson(x, y)
print(f"Pearson's Correlation Coefficient (r): {r_value:.4f}")

```

## Visualization

Linear relationship analysis with trendline showing the correlation between X and Y coordinates.

```python
# Create visualization
plt.figure(figsize=(10, 6))

# Plot data points
plt.scatter(x, y, color='#1f77b4', s=100, label='Data Points', edgecolors='black', zorder=3)

# Calculate and plot trendline
m, b = np.polyfit(x, y, 1)
x_line = np.array([min(x), max(x)])
y_line = m * x_line + b
plt.plot(x_line, y_line, color='red', linestyle='--', linewidth=2, label=f'Trendline (r={r_value:.4f})')

# Customize plot
plt.title("Linear Relationship Analysis (AIML Midterm)", fontsize=14, pad=20, fontweight='bold')
plt.xlabel("X Coordinate", fontsize=12)
plt.ylabel("Y Coordinate", fontsize=12)
plt.axhline(0, color='black', linewidth=0.8, linestyle='-', alpha=0.3)
plt.axvline(0, color='black', linewidth=0.8, linestyle='-', alpha=0.3)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(fontsize=11, loc='upper left')

plt.tight_layout()
plt.show()

# Print regression equation
print(f"\nRegression Equation:")
print(f"y = {m:.4f}x + {b:.4f}")
```

## Summary Statistics

Additional analysis of the data distribution and relationship strength.

```python
# Calculate additional statistics
mean_x = mean(x)
mean_y = mean(y)
std_x = sigma(x, mean_x) / len(x)
std_y = sigma(y, mean_y) / len(y)
r_squared = r_value ** 2

print("Statistical Summary:")
print("="*50)
print(f"X Data:")
print(f"  Mean: {mean_x:.4f}")
print(f"  Min: {min(x):.4f}")
print(f"  Max: {max(x):.4f}")
print(f"\nY Data:")
print(f"  Mean: {mean_y:.4f}")
print(f"  Min: {min(y):.4f}")
print(f"  Max: {max(y):.4f}")
print(f"\nCorrelation Analysis:")
print(f"  Pearson's r: {r_value:.4f}")
print(f"  R-squared (rÂ²): {r_squared:.4f}")
print(f"  Slope (m): {m:.4f}")
print(f"  Intercept (b): {b:.4f}")
print("="*50)
```
